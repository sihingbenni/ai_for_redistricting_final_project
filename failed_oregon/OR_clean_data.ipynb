```python
def test_render(gdf: gpd.GeoDataFrame, title):
 # Dissolve geometries by SEND so each district is represented once
    gdf_dissolved = gdf.dissolve(by="SHORTNAME", as_index=False)

    # Generate N distinct colors
    def get_n_colors(n):
        colors = plt.cm.get_cmap('hsv', n)
        return [colors(i) for i in range(n)]

    # Generate unique colors for each SEND
    districts = gdf_dissolved["SHORTNAME"].unique()
    color_list = get_n_colors(len(districts))
    color_map = {district: color_list[i] for i, district in enumerate(districts)}

    # Map colors back to original gdf
    gdf["color"] = gdf["SHORTNAME"].map(color_map)

    # Plot original geometries
    ax = gdf.plot(color=gdf["color"], figsize=(20, 20), legend=True)

    # Plot one dot at the centroid of each dissolved (unique) SEND geometry
    gdf_dissolved["geometry"].centroid.plot(ax=ax, color="white", markersize=50)

    plt.axis("off")
    plt.title(title, fontsize=82)
    plt.tight_layout()
    plt.savefig(f"images/oregon_map_{title}.svg", bbox_inches="tight")
    plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt
import shapely
import pandas as pd

def highlight_overlaps_gdf(gdf: gpd.GeoDataFrame, output_svg: str = "overlaps.svg"):
    # Copy and fix invalid geometries with buffer(0)
    gdf = gdf.copy()
    gdf["geometry"] = gdf["geometry"].buffer(0)

    # Use spatial index
    overlaps_list = []
    sindex = gdf.sindex

    for idx, geom in gdf.geometry.items():
        if not geom.is_valid:
            continue  # skip invalid geometry

        possible_matches_index = list(sindex.intersection(geom.bounds))
        possible_matches = gdf.iloc[possible_matches_index]

        for other_idx, other_geom in possible_matches.geometry.items():
            if idx >= other_idx or not other_geom.is_valid:
                continue

            try:
                if geom.intersects(other_geom):
                    intersection = geom.intersection(other_geom)
                    if not intersection.is_empty:
                        overlaps_list.append(intersection)
            except shapely.errors.GEOSException as e:
                print(f"Skipped invalid geometry pair ({idx}, {other_idx}): {e}")
                continue

    # Convert to GeoSeries
    overlaps = gpd.GeoSeries(overlaps_list)

    # Plot base map
    ax = gdf.plot(facecolor="lightgray", edgecolor="black", figsize=(12, 12))

    # Plot overlaps
    if not overlaps.empty:
        overlaps.plot(ax=ax, color="red")

    # Final touches
    plt.axis("off")
    plt.tight_layout()
    plt.savefig(output_svg, format="svg", bbox_inches="tight")
    plt.close()
    print(f"✅ PNG with overlaps saved to: {output_svg}")
```
