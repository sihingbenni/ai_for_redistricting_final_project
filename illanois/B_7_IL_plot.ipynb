# Outlier Analysis/Histograms

```python
import matplotlib.pyplot as plt
import pickle
```

```python
# Load the data from the .pkl file
with open('checkpoints/IL_plot_results_35_000.pkl', 'rb') as file:
    checkpoint_dict = pickle.load(file)
```

```python
# Extract the cut edge ensemble
cutedge_ensemble = checkpoint_dict["list_of_nr_of_cut_edges"]
print(cutedge_ensemble)
```

```python
cutedge_ensemble_half = checkpoint_dict["list_of_nr_of_cut_edges"][:len(checkpoint_dict["list_of_nr_of_cut_edges"]) // 2]
```

```python
# Plot the histogram
plt.figure()
plt.hist(cutedge_ensemble_half, align='left', bins=30)  # Adjust bins as needed
plt.title("Histogram of Cut Edges")
plt.xlabel("Number of Cut Edges")
plt.ylabel("Frequency")
plt.show()
```

```python
# Plot the histogram
plt.figure()
plt.hist(cutedge_ensemble, align='left', bins=30)  # Adjust bins as needed
plt.title("Histogram of Cut Edges")
plt.xlabel("Number of Cut Edges")
plt.ylabel("Frequency")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Overlay the histograms
plt.figure()
plt.hist(cutedge_ensemble_half, bins=30, alpha=0.5, label="First Half", color="blue", align='left')  # First half
plt.hist(cutedge_ensemble, bins=30, alpha=0.5, label="Full Ensemble", color="orange", align='left')  # Full ensemble
plt.title("Comparison of Cut Edges")
plt.xlabel("Number of Cut Edges")
plt.ylabel("Frequency")
plt.legend()  # Add a legend to distinguish the datasets
plt.show()
```
